Q5. In a machine learning project, you have a dataset with 1000 rows and 5 columns. Two of the columns

```python
import pandas as pd
```

```python
df=pd.read_csv('titanic.csv')
df.head()
```

```python
s=df[['PassengerId','Survived','Pclass','Name','Sex']]
```

```python
s
```

```python
# now we have to use nominal to transform the categorical data
from sklearn.preprocessing import OneHotEncoder
encoder=OneHotEncoder()
```

```python
encoder.fit_transform(df[['Name','Sex']]).toarray()
```

```python
encoded=encoder.fit_transform(df[['Name','Sex']])
```

```python
import pandas as pd
encoded_data=pd.DataFrame(encoded.toarray(),columns=encoder.get_feature_names_out())
```

```python
encoded_data
```
